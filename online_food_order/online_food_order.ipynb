```python
gender_counts = df['Gender'].value_counts()
gender_percentage = (gender_counts / gender_counts.sum() * 100).round(2)
```

```python
print("Gender count:")
print(gender_counts)
print("\nGender percentage:")
print(gender_percentage)
gender_percentage.plot(kind='bar', color=['blue', 'pink'], width=0.2)

plt.title('Gender percentage')
plt.xlabel('Gender')
plt.ylabel('percentage (%)')
plt.xticks(rotation=0)

plt.show()
```

```python
marital_status_counts = df['Marital Status'].value_counts()
marital_status_percentage = (marital_status_counts / marital_status_counts.sum() * 100).round(2)
print("Marital Status count:")
print(marital_status_counts)
print("\nMarital Status percentage:")
print(marital_status_percentage)
marital_status_percentage.plot(kind='bar', color=['blue', 'red', 'grey'], width=0.2)

plt.title('Marital Status percentage')
plt.xlabel('Marital Status')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)

plt.show()
```

```python
occupation_counts = df['Occupation'].value_counts()
occupation_percentage = (occupation_counts / occupation_counts.sum() * 100).round(2)
print("Occupation count:")
print(occupation_counts)
print("\nOccupation percentage:")
print(occupation_percentage)
occupation_percentage.plot(kind='bar', color=['blue', 'green', 'grey'], width=0.2)

plt.title('Occupation percentage')
plt.xlabel('Occupation')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0) 

plt.show()
```

```python
monthly_income = df['Monthly Income'].value_counts()
monthly_income_percentage = (monthly_income / monthly_income.sum() * 100).round(2)

print("Monthly Income count:")
print(monthly_income)

print("\nMonthly Income percentage:")
print(monthly_income_percentage)

new_labels = [label.replace(' ', '\n') for label in monthly_income.index]
unique_categories = monthly_income.index.nunique()
colors = plt.cm.viridis(np.linspace(0, 1, unique_categories))

monthly_income_percentage.plot(kind='bar', color=colors, width=0.2)

plt.title('Monthly Income percentage')
plt.xlabel('Monthly Income')
plt.ylabel('Percentage (%)')
plt.xticks(range(len(new_labels)), new_labels, rotation=45)
plt.tight_layout()

plt.show()
```

```python
educational_qualifications = df['Educational Qualifications'].value_counts()
educational_qualifications_percentage = (educational_qualifications / educational_qualifications.sum() * 100).round(2)

print("Educational Qualifications Income count:")
print(educational_qualifications)

print("\nEducational Qualifications percentage:")
print(educational_qualifications_percentage)

new_labels = [label.replace(' ', '\n') for label in educational_qualifications.index]
unique_categories = educational_qualifications.index.nunique()
colors = plt.cm.viridis(np.linspace(0, 1, unique_categories))

educational_qualifications_percentage.plot(kind='bar', color=colors, width=0.2)

plt.title('Educational Qualifications Income percentage')
plt.xlabel('Educational Qualifications Income')
plt.ylabel('Percentage (%)')
plt.xticks(range(len(new_labels)), new_labels, rotation=45)
plt.tight_layout()

plt.show()
```
